## JSON

```python
d = {"name": "혼자 공부하는 데이터 분석"}

print(d['name'])
```

```python
import json
```

```python
d_str = json.dumps(d, ensure_ascii=False)
print(d_str)
```

```python
print(type(d_str))
```

```python
d2 = json.loads(d_str)

print(d2['name'])
```

```python
print(type(d2))
```

```python
d3 = json.loads('{"name": "혼자 공부하는 데이터 분석", "author": "박해선", "year": 2021}')

print(d3['name'])
print(d3['author'])
print(d3['year'])
```

## XML

```python
x_str = """
<book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2021</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et

book = et.fromstring(x_str)
```

```python
print(type(book))
```

```python
print(book.tag)
```

```python
book_childs = list(book)

print(book_childs)
```

```python
name, author, year = book_childs

print(name.text)
print(author.text)
print(year.text)
```

```python
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')

print(name)
print(author)
print(year)
```

```python
x2_str = """
<books>
    <book>
        <name>혼자 공부하는 데이터 분석</name>
        <author>박해선</author>
        <year>2021</year>
    </book>
    <book>
        <name>혼자 공부하는 머신러닝+딥러닝</name>
        <author>박해선</author>
        <year>2020</year>
    </book>
</books>
"""
```

```python
books = et.fromstring(x2_str)

print(books.tag)
```

```python
for book in books.findall('book'):
    name = book.findtext('name')
    author = book.findtext('author')
    year = book.findtext('year')
    
    print(name)
    print(author)
    print(year)
    print()
```
